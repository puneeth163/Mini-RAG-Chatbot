```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import os
import warnings

# Setup to keep the output clean
os.environ["TOKENIZERS_PARALLELISM"] = "false"
warnings.filterwarnings("ignore")

# 1. Step: Create Knowledge Base
# This serves as the 'authoritative' data your model will reference.
documents = [
    "Machine Learning is a subset of artificial intelligence that enables systems to learn from data.",
    "RAG (Retrieval-Augmented Generation) combines retrieval with generative AI models.",
    "Python is the most popular language for Data Science and AI development.",
    "Vector databases store information as high-dimensional embeddings for fast retrieval.",
    "Large Language Models (LLMs) generate human-like text based on input prompts."
]

# 2. Step: Load Embedding Model
# We use a lightweight pre-trained model from Hugging Face.
print("Loading embedding model...")
model = SentenceTransformer('all-MiniLM-L6-v2')

# 3. Step: Create Embeddings
# Convert your text documents into numerical vectors.
doc_embeddings = model.encode(documents)
print("Documents converted into embeddings.")
print(f"Embedding shape: {doc_embeddings.shape}")
print("\nRAG Chatbot is Ready!\nType 'exit' to stop.\n")

# 4. Step: Build the RAG Logic
def chatbot():
    while True:
        query = input("Ask a question: ")
        if query.lower() == 'exit':
            print("Stopping the RAG system...")
            break
        
        # Retrieval: Convert query to vector and find the best match
        query_embedding = model.encode([query])
        similarities = cosine_similarity(query_embedding, doc_embeddings)
        best_idx = np.argmax(similarities)
        
        # Display the most relevant document as the "Answer"
        print(f"Answer: {documents[best_idx]}\n")

if __name__ == "__main__":
    chatbot()
```
